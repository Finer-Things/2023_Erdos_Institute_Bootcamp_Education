```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Getting the two Regents Excel Files

```python
are_df = pd.read_csv("../Data/SRC2022/Annual Regents Exams Toy Example.csv")
```

```python
are_df["PER_LEVEL1"].value_counts()
```

```python
df = are_df[are_df["YEAR"] == 2021]
# df = df[are_df["SUBGROUP_NAME"].isin(["Male", "Female"])].dropna(subset=["PER_LEVEL1", "PER_LEVEL2"])
df = df[pd.to_numeric(df['PER_LEVEL1'], errors='coerce').notnull()][pd.to_numeric(df['PER_LEVEL2'], errors='coerce').notnull()][are_df["SUBGROUP_NAME"].isin(["Male", "Female"])]

for column in ['PER_LEVEL1', 'PER_LEVEL2']:
    df[column] = df[column].astype("int64")

sns.scatterplot(data=df, x="NUM_LEVEL1", y="NUM_LEVEL2", hue="SUBGROUP_NAME", )
```

```python
df["PER_LEVEL1"].describe()
```

```python
sns.pairplot(data=df, x_vars=['PER_LEVEL1', 'PER_LEVEL2', "PER_LEVEL3"], y_vars=['PER_LEVEL1', 'PER_LEVEL2'], hue="SUBGROUP_NAME")
```
